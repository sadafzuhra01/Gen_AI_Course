## Inheritence wrt OOP
- Single Inheritence
- Multiple Inheritence

```python
class Car:
    ## attributes are defined by using constructor __init__
    def __init__(self,windows,doors,engine_type):
        self.windows= windows
        self.doors= doors
        self.engine_type=engine_type

    def drive(self):
        print(f"The person drives a {self.engine_type} car")



```

```python
audi = Car(2,2,"petrol")
```

```python
audi.drive()
print(audi.windows)
```

## 1- Single Inheritence

```python
class Tesla(Car):
    def __init__(self,windows,doors,engine_type,is_self_driving):
        super().__init__(windows,doors,engine_type)
        self.is_self_driving = is_self_driving

    def self_drive(self):
        print(f"This car has self driving functionality: {self.is_self_driving}")
    


```

```python
tesla1 = Tesla(4,4,"electric",True)
```

```python
print(tesla1.engine_type)
tesla1.self_drive()
tesla1.drive()
```

## 2- Multiple Inheritence

```python
## Class 1


class Animal:
    def __init__(self,name):
        self.name= name

    def walk(self):
        print(f"The {self.name} is walking on the ground.")


animal1 = Animal("Louis Pasteure")
animal1.walk()
        

```

```python
## Class 2

class Pet:
    def __init__(self,owner):
        self.owner=owner
```

```python
class Dog(Animal, Pet):
    def __init__(self, name, owner, colour):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        self.colour=colour

    def speak(self):
        print(f"{self.name} says woof!")
```

```python
dog1 = Dog("Bunny", "Sadia", "brown")
```
