```python
## Encapsulation

class Animal:
    def __init__(self, name, age):
        self.name = name ## public variables
        self.age = age   ## public variables
    
    def speak(self):
        print("Animal makes a sound")


dog = Animal("jackie", 3)



    
```

```python
class Animal:
    def __init__(self, name, age):
        self.__name = name ## private variables
        self.__age = age   ## private variables
    
    def speak(self):
        print("Animal makes a sound")


dog1 = Animal("tomi", 2)

```

```python
## protected variables
class Animal:
    def __init__(self, name, age):
        self._name = name ## protected variables
        self._age = age   ## protected variables
    
    def speak(self):
        print("Animal makes a sound")


dog2 = Animal("stark", 1)




```

```python
dir(dog2)
```

```python
class Animal:
    def __init__(self, name, age):
        self.__name = name ## private variables
        self.__age = age   ## private variables
    
    def speak(self):
        print("Animal makes a sound")
    
    ## getter and setter method

    def get_name(self):
        return self.__name
    
    def set_age(self, age):
        if age > 0:
            self.__age = age

    def get_age(self):
        return self.__age

    


dog4 = Animal('Max', '3')

print(dog4.get_name())
print(dog4.set_age(4))
print(dog4.get_age())

    


```

### Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Vehicle(ABC):
    def __init__(self, company):
        self.company = company


    def drive(self):
        print(f"The person drives {self.company} vehicle. ")

    @abstractmethod
    def start_engine(self):
        pass



class Car(Vehicle):
    def __init__(self, company):
        super().__init__(company)
    
    def start_engine(self):
        return f"the {self.company} engine started."
    


car1 = Car('Audi')

    
print(car1)


    
```

### Operator Overloading

```python
## __add__(self, other), +
## __sub__(self, other), -
## __mul__(self, other), *
## __truediv__(self, other), /
```

```python
class Vector:
    def __init__(self, x,y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __repr__(self):
        return f"Vector {self.x}, {self.y}"
    

v1 = Vector(3,6)
v2 = Vector(4,5)

print(v1+v2)
print(v1-v2)
```
