```python
## Real World Examples 

## Managing a TO DO list

tasks = ['Namaz', 'Running', 'University', 'Meeting YPDC', 'Zohr', 'Lunch', 'Back to home']

##

tasks.append('Revise Lecture')
tasks.append('Meeting with friends')
print(tasks)
## 
tasks.remove('Namaz')
tasks.remove('Running')
print(tasks)
## 

for task in tasks:
    if task in tasks:
        print(f"Don't forget to perform this {task}")
    
```

```python
## Inventory Management

invent = ['apples', 'bananas', 'cherry', 'Mangoes', 'Grapes']

##
invent.append('watermelon')
invent.append('pineapple')
print(invent)

## 
invent.remove('bananas')
print(invent)

##

for fruit in invent:
    if fruit in invent:
        print(f"{fruit} is available in the stock")
    else:
        print(f"{fruit} is not available in the stock")


```

```python
print(invent)

## iterate and traverse

print(invent[0])

for i in invent:
    print(f"{i}, is a fruit")
    
        
    
```

```python
## Counting the FeedBack

feedback = ["Good Service", "Great Service", "Excellent Service", "Worse Service"]

feedback.append("Ordinary Servivce")
feedback.append("Good Service")
feedback.append("Great Service")
feedback.append("Excellent Service")

## Positive Feedback

positive_feedback = sum(1 for comment in feedback if 'good' in comment.lower() or 'great' in comment.lower() or 'excellent' in comment.lower())

print(f"This the number {positive_feedback} we recieved the positive feedback")
```
