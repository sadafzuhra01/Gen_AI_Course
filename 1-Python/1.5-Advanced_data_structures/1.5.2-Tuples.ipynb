```python
## Creating tuples
tpl = ()
print(type(tpl))
```

```python
tpl = ("haseeb", "Raza", "Mubeen", 0,1,2,3,4,5,6, True, False)
print(type(tpl))
```

```python
## Accessing data 

print(tpl[0])
print(tpl[1])
print(tpl[7])

## Slicing operations 
```

```python
numbers = (1,2,3,4,5,6,7)
print(numbers[0:5]) ## Slicing
```

```python
## Tuple Operations
## concatenation
## Multiplication
print(tpl + numbers) ## concatenation
print(tpl * 3)
```

```python
## Packing and Unpacking
name_v1 = 1,2,3,4,5,6
name = ("Haseeb", "Raza", "DS")
a,b,c = name


print(type(name_v1))
print(a)
print(b)
print(c)
```

```python
## Immutable Nature 

numbers = (1,2,3,4,5,6,7,8)

numbers[2] = "Haseeb"
```

```python
## Nested List/Tuples
lst = [[1,2,3,4,5], ["a","b","c","d","e"],[True,False,True,True,False]]

tpl = ((1,2,3,4,5), ("a","b","c","d","e"),(True,False,True,True,False))
```

```python
## Iterating over the Nested Tuple

lst=[]

for sub_tpl in tpl:
    for i in sub_tpl:
        if type(i)==int:
            lst.append(i*21)
        else:
            continue


print(lst)
        
```
