```python
## File Operations
## modes: w, r, a, w+, wb, and rb

with open('example.txt', 'w') as file:
    file.write("Hello, admin!")
    file.write("\nThis is the second line.")
```

```python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
with open('example.txt', 'w') as file:
    file.write('Hello, world!')


with open('example.txt', 'r') as file:
    content = file.read()
    print(content) 
```

```python
with open('example.txt', 'a') as file:
    file.write('\nHello, admin')

with open('example.txt', 'r') as file:
    print(file.read())
```

```python
with open('example.txt', 'w+') as file:
    file.write('Hello, world!')
    file.write('\nhello, admin')
    file.write('\nhaseeb is a data scientist.')

    file.seek(0)

    content = file.read()
    print(content)

```
