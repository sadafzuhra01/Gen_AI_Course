```python
## While Loop

count=0

while count<5:
    print(count)
    count = count+1
```

```python
n=100
sum=0

while n<=1100:
    sum = sum + n
    n = n+1

print(sum)
```

```python
## Factorial Calculation using while loop
## factorial of 5

n=100
fact=1

while n>=1:
    fact=fact*n
    n=n-1


print(fact)
```

```python
## Loop Control Statements

## Break Statement
n=100
fact=1

while n>=1:
    if n==50:
        break
    fact=fact*n
    n=n-1


print(fact)
```

```python
## Continue statement : skip 


## printing odd numbers from 0 to 100
for i in range(100):
    if i % 2 == 0:
        continue
    print(i, end=' ')
```

```python
## pass statement: Null statement
for i in range(100):
    if i % 2 == 0:
        pass
    print(i, end=' ')
```

```python
## Nested Loops
## loop inside a loop

for i in range(3):
    for j in range(3):
        print(f"({i}, {j})")

```
