```python

age = 16
if age>=18:
    print("You are eligible for voting")

else:
    print("You are not eligible for voting")
```

```python
age =21

if age<=12:
    print("You are a child")

elif age<=19:
    print("You are a teenager")

elif age<=30:
    print("You are an adult")

else:
    print("You are an elder")
```

```python
## if a number is even, odd, or negative
## Nested if-else

num = int(input("Enter a number:")) 

if num>0:
    print(f"{num},is a positive number.")
    if num%2==0:
        print(num,"is even")
    else:
        print(num,"is odd")
elif num==0:
    print(num,"is zero")
else:
    print(num,"is negative")
    
```

```python
## simple calculator 

num_1 = int(input("Enter the first number:"))
num_2 = int(input("Enter the second number:"))

operation= input("Operation to be performed:(+,-,/,*)")

if operation=="+":
    print(num_1 + num_2)
elif operation=="-":
    print(num_1 - num_2)
elif operation=="/":
    if num_2 != 0:
        print(num_1 / num_2)
    else:
        print("Error! Division by zero is not allowed")
elif operation=="*":
    print(num_1 * num_2)
else:
    print("invalid operation, please enter the right one.")
```
