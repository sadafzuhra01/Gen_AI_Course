```python
def even_or_odd(num):
    """this function distinguish between even and odd numbers"""
    if num % 2 != 0:
        result = "Odd Number"
    else:
        result = "Even Number"
    
    return result
```

```python
even_or_odd(54)
```

```python
## Default Parameter
def greet(name="Mubeen"):
    """Prints a greeting message with the given name"""
    if name=="Mubeen":
        print(f"{name}, get out of this store!")
    else:
        print(f"Aoa!, {name}!")

```

```python
greet()
```

```python
## Parameters == Arguments

## Variable Length Arguments 
##1- Positional Arguments
##2- Keyword Arguments

def print_argument(*args,**kwargs):
    for i in args:
        print(f"{i},how have you been?")
    
    for key,value in kwargs.items():
        print(f"{key} is {value}")
```

```python
print_argument("haseeb","Mubeen","tayyaba","Furmein",name="Sadia",age=21)
```
