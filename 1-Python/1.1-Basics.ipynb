```python
## Simple Calculator 
## input() function


num_1 = float(input("Write the first number: "))
num_2 = float(input("Write the second number: "))


sum = num_1 + num_2
mult = num_1 * num_2
sub = num_1 - num_2
div = num_1 / num_2




print(f"Sum of the above two numbers is: {sum}.")
print(f"Difference of the above two numbers is: {sub}.")
print(f"Multiplication of the above two numbers is: {mult}.")
print(f"Division of the above two numbers is: {div}.")
```
