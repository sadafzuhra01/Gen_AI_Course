#### 1- Arithematic Operators
Arithematic operators are used to perform basic mathematical operations like addition, subtraction, multiplication, division.

```python
## +,-,*,/,//,%

num_1 = 21
num_2 = 7

sum = num_1 + num_2
diff = num_1 - num_2
mult = num_1 * num_2
div = num_1 / num_2
floor_div = num_1 // num_2 ## Floor Division = near lowest number without decimal
mod = num_1 % num_2        ## Modulus = remainder


print(sum)
print(diff)
print(mult)
print(div)
print(floor_div)
print(mod) 
```

#### 2- Comparison Operators 
The comparison operators are used to compare two values. The comparison operators are:
- Equal to (==)
- "<"
- ">"
- "<="
- ">="
- "!="
  

```python
a= 21 
b= 20
print(a==b)
```

```python
print( a>b)
```

```python
print(a<b)
print(a>=b)
print(a<=b)
print(a!=b)
```

#### 3- Logical Operators 
Logical operators are used to combine multiple conditions in a single statement. The three logical operators are:
- and
- or
- not

```python
x= True
y= False

x and y
```

```python
x= True
y= False

x or y
```

```python
x = True

not x
```
