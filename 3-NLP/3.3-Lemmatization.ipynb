## Lemmatization with Wordnet Lemmatizer

```python
## Stemming ----> word stem
## Lemmatization ----> word lemma 
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()

lemmatizer.lemmatize('going')
```

```python
'''
pos tags ---> noun = n
pos tags ---> verb = v
pos tags ---> adverb = r
pos tags ---> adjective = a
'''
```

```python
lemmatizer.lemmatize('going', pos = 'v' )
```

```python
words = ['sportingly', 'fairly', 'fastly', 'slowly', 'hardly' ]

for word in words:
    print(word+"----->"+lemmatizer.lemmatize(word, pos ='n'))
```

```python
words = ['eaten', 'eats', 'eating', 'writes', 'writing', 'written', 'going', 'goes']
for word in words:
    print(word+"----->"+lemmatizer.lemmatize(word, pos ='v'))
```
