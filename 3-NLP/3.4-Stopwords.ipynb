## Stopwords in NLP

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
corpus = '''When he was ten years old, Huang lived with his brother in the Oneida boys' dormitory.[14] Each student was expected to work every day, and his older brother was assigned to perform manual labor on a nearby tobacco farm.[15] Because he was too young to attend classes at the reform academy, Huang was educated at a separate public school—the Oneida Elementary school in Oneida, Kentucky—arriving as "an undersized Asian immigrant with long hair and heavily accented English"[14] and was frequently bullied and beaten.[16] In Oneida, Huang cleaned toilets every day, learned to play table-tennis,[b] joined the swimming team,[18] and appeared in Sports Illustrated at age 14.[19] He taught his illiterate roommate, a "17-year-old covered in tattoos and knife scars,"[19] how to read in exchange for being taught how to bench press.[14] In 2002, Huang recalled that he remembered his life in Kentucky "more vividly than just about any other".[19]
'''
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)

for sentence in sentences:
    print(sentence)
```

```python
## for loop using range function
## sentences ---> tokenization
## words ----> stopwords
## stemming/lemmatization
## joining words to sentences 
```

```python
from nltk.tokenize import word_tokenize
```

```python
stopwords.words('english')
```

```python
stemmer = PorterStemmer()
```

```python

for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)

```

```python
for sentence in sentences:
    print(sentence)
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowstemmer = SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [snowstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)

for sentence in sentences:
    print(sentence)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word.lower(), pos ='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)

for sentence in sentences:
    print(sentence)
```
