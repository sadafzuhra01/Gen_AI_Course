## Tokenization with NLTK

```python
## Tokenization is a component of text preprocessing - (data/text cleaning)
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
```

```python
## Corpus
## Documents
## Words
## Vocabulary

corpus = """Haseeb is a data scientits. He is leading youth in AI industry, his university has 
rewarded him with some gifts. Soon, he will be representing UAF, at Ministry of Planning Development and 
special initiatives."""




print(corpus)
```

```python
## Practical - Tokenization
## Tokenization ----> Corpus/Sentences ---> sentences/words

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)
print(sentences)
```

```python
for sentence in sentences:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(corpus)

for word in words:
    print(word)
```

```python
from nltk.tokenize import wordpunct_tokenize

```

```python
words = wordpunct_tokenize(corpus)

for word in words:
    print(word)
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

dir(tokenizer)
```

```python
words = tokenizer.tokenize(corpus)

for word in words:
    print(word)
```
