## Stemming - Text Processing Technique

```python
## Word (Suffixes/Prefixes) ---> root/lemma
```

```python
## 1- Porter Stemmer
## 2- Regexp Stemmer
## 3- Snowball Stemmer
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
words = ['eating','eatable', 'goes', 'going', 'go', 'writing', 'writes', 'wrote', 'written','history','edible' ]
```

```python
stemmer = PorterStemmer()
```

```python
dir(stemmer)
```

```python
for word in words:
    print(word+"----->"+stemmer.stem(word))
```

```python
## Regexp Stemmer 

from nltk.stem import RegexpStemmer
```

```python
regstemmer = RegexpStemmer('ing$|s$|es$|able$', min=4)
dir(regstemmer)
```

```python
for word in words:
    print(word+"----->"+regstemmer.stem(word))
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowstemmer = SnowballStemmer('english')

```

```python
for word in words:
    print(word+"----->"+snowstemmer.stem(word))
```

```python
print(stemmer.stem('Hideous'))
print(snowstemmer.stem('Hideous'))
```
